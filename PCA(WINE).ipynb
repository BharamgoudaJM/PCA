# HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage
```

```python
a=linkage(final,method='complete',metric='euclidean')
```

```python
plt.figure(figsize=(12,8))
plt.title("Dendrogram")
dendrogram=sch.dendrogram(sch.linkage(final,method='complete'))
```

```python
complete=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity='euclidean').fit(final)
```

```python
complete.labels_
```

```python
c_labels=pd.Series(complete.labels_)
```

```python
final['cluster']=c_labels
final
```

```python
final['cluster'].value_counts()
```

```python
complete.fit(final)
```

```python
from sklearn.metrics import silhouette_score as sil
```

```python
cc=[3,4,5,6,7,8,9]
score=[]
for i in cc:
    hc=AgglomerativeClustering(n_clusters=i,affinity='euclidean',linkage='complete')
    z=hc.fit_predict(final.iloc[:,[1,2,3]])
    score.append(sil(final.iloc[:,[1,2,3]],z))
```

```python
score
```

```python
plt.plot(cc,score,color="purple")
plt.scatter(cc,score,color="black")
plt.show()
```

# K MEANS

```python
from sklearn.cluster import KMeans
```

```python
tt=[]
k=list(range(1,8))

for i in k:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(final)
    tt.append(kmeans.inertia_)
```

```python
tt
plt.plot(k,tt,'ro-')
plt.xlabel("number of clusters")
plt.ylabel("total")
```

```python
model1=KMeans(n_clusters=3)
model1.fit(final)
```

```python
model1.labels_
mj=pd.Series(model1.labels_)
final['kmeans']=mj
```
